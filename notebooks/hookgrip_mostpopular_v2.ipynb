```python
def count_compete(tagged_list, name_list):    
    year_compete = []
    year_list = [['2023', ''], ['2022', ''], ['2021', ''], ['2020', ''], ['2019', ''], ['2018', ''], ['2017', ''], ['2016', ''],\
             ['2015', ''], ['2014', ''], ['2013', '']]
    for item in data:
        if 'mentions' in item:
            for name in item['mentions'] :
                if name in tagged_list:
                    year_compete.append(item['timestamp'][0:4])
                else:
                    for name in name_list:
                        if ('snatch' or 'clean' or 'jerk') in item['caption']\
                        and name in item['caption']:
                            year_compete.append(item['timestamp'][0:4])
        
    for year in year_list:
        year[1] = year_compete.count(year[0])

    return year_list

kuo2 = count_compete(kuo_hs, ['Kuo Hsingchun'])
print(kuo2)

lasha2 = count_compete(lasha, ['Lasha Talakhadze'])

counter = 0
for num in lasha2:
    counter += num[1]

print(counter)
```
